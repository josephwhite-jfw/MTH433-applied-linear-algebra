```python
import numpy as np
import pandas as pd

def matrix_prep(A):
    """
    """
    # Standardize (Demean)
    mu = np.mean(A, axis=0)
    A  = A - mu

    # Normalize
    A /= np.linalg.norm(A, axis=0)
    return A
```

```python
def pc_var(A,k):
    U, s, Vt = np.linalg.svd(A, full_matrices=False)
    V = Vt.T
    # Get the V coordinates 
    V_coordinates = A @ V
    
    V_coordinates_k = V_coordinates[:, :k]
    
    s2 = s**2
    cumulative_var = np.cumsum(s2) / np.sum(s2)
    return V_coordinates_k, cumulative_var
```
