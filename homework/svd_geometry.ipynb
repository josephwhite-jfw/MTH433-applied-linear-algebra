$\huge A = \begin{bmatrix} 2 & 1 \\ 1 & 1 \end{bmatrix}$

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 1],
              [1, 1]])
U, S, Vh = np.linalg.svd(A)

print("U =\n", U)
print("Singular values S =", S)
print("V^T =\n", Vh)

# Break down A = U S V^T into three separate actions.

t = np.linspace(0, 1, num=100)
circ_x = np.cos(2 * np.pi * t)
circ_y = np.sin(2 * np.pi * t)

# 1) Unit circle and singular vectors
plt.plot(circ_x, circ_y, color='k')
plt.plot(t * Vh[0, 0], t * Vh[0, 1])
plt.plot(t * Vh[1, 0], t * Vh[1, 1])
plt.axis('equal')

# 2) Applying V^T
plt.figure()
plt.plot(circ_x, circ_y, color='k')
plt.plot(t * 1, t * 0)
plt.plot(t * 0, t * 1)
plt.axis('equal')

# 3) Applying S
plt.figure()
plt.plot(S[0] * circ_x, S[1] * circ_y, color='k')
plt.plot(t * S[0], t * 0)
plt.plot(t * 0, t * S[1])
plt.axis('equal')

# 4) Applying U
plt.figure()
x_final = U[0, 0] * S[0] * circ_x + U[0, 1] * S[1] * circ_y
y_final = U[1, 0] * S[0] * circ_x + U[1, 1] * S[1] * circ_y
plt.plot(x_final, y_final, color='k')
plt.plot(t * S[0] * U[0, 0], t * S[0] * U[1, 0])
plt.plot(t * S[1] * U[0, 1], t * S[1] * U[1, 1])
plt.axis('equal')
plt.title("SVD geometry for A")
```
