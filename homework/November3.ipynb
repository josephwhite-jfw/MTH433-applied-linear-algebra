```python
import numpy as np
import matplotlib.pyplot as plt
```

$\huge A = \left[  \begin{array}{cc} 1 & 1\\ 0 & 1 \end{array}  \right] $

```python
A = np.array([[1,1],[0,1]])
```

```python
U,S,Vh = np.linalg.svd(A)
```

$$ \huge U = \begin{bmatrix}
0.85065081 & -0.52573111 \\
0.52573111 & 0.85065081
\end{bmatrix}, \quad S = \begin{bmatrix}
1.61803399 & 0 \\
0 & 0.61803399
\end{bmatrix}, \quad V^T = \begin{bmatrix}
0.52573111 & 0.85065081 \\
-0.85065081 & 0.52573111
\end{bmatrix} $$

# We break down $ A = USV^T$ into three separate actions.

```python
t = np.linspace(0,1,num=100);
circ_x = np.cos(2*np.pi*t);
circ_y = np.sin(2*np.pi*t);
plt.plot(circ_x,circ_y,color='k');
plt.plot(t*Vh[0,0],t*Vh[0,1]);
plt.plot(t*Vh[1,0],t*Vh[1,1]);
plt.axis('equal'); #to make sure it actually looks like a circle
```

# Columns of V $\uparrow$. Multiply by $V^T$ to get $\downarrow$

```python
plt.plot(circ_x,circ_y,color='k');
plt.plot(t*1,t*0);
plt.plot(t*0,t*1);
plt.axis('equal');
```

# Multiply by S to get $\downarrow$

```python
plt.plot(S[0]*circ_x,S[1]*circ_y,color='k');
plt.plot(t*S[0],t*0);
plt.plot(t*0,t*S[1]);
plt.axis('equal'); #to make sure it looks like an ellipse
```

# Finally apply U to see what A = USV.T does to the unit circle $\downarrow$

```python
plt.plot(U[0,0]*S[0]*circ_x+U[0,1]*S[1]*circ_y,U[1,0]*S[0]*circ_x+U[1,1]*S[1]*circ_y,color='k');
plt.plot(t*S[0]*U[0,0],t*S[0]*U[1,0]);
plt.plot(t*S[1]*U[0,1],t*S[1]*U[1,1]);
plt.axis('equal');
```
